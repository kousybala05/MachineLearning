### MATPLOTLIB

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("auto-mpg.csv")
```

```python
df.head()
```

```python
print(plt.style.available)
```

```python
#lineplot--->relation between two variables
```

```python
plt.plot('displacement','weight',data=df,color='g',marker='*')
plt.show()
```

```python
df1=df.head(20)
```

```python
plt.plot('mpg','horsepower',data=df1,color='y',marker='o',markersize=5)

plt.title("LINEPLOT")
plt.xlabel('mpg')
plt.ylabel('horsepower')

plt.show()
```

```python
plt.plot('mpg','horsepower',data=df1,color='y',marker='o',markersize=5)
plt.plot('acceleration','weight',data=df1,color='g',marker='o',markersize=5)
plt.title("LINEPLOT")
plt.xlabel('mpg')
plt.ylabel('horsepower')
plt.grid()
plt.show()
```

```python
plt.plot('mpg','horsepower',data=df1,color='k',marker='o',markersize=5)

plt.title("LINEPLOT")
plt.xlabel('mpg')
plt.ylabel('horsepower')
plt.grid(linestyle='-',linewidth='6',color='g')
plt.show()
```

###seaborn

```python
import seaborn as sns
```

```python
sns.stripplot(y='weight',x='car name',data=df1)
plt.show()
```

```python
dfnew=pd.read_csv('Automobile.csv')
dfnew
```

```python
#strip plot
sns.stripplot(x='wheel_base',y='number_of_doors',data=dfnew)
plt.show()
```

```python
sns.stripplot(x='wheel_base',y='number_of_doors',hue='fuel_type',data=dfnew)
plt.show()
```

```python
#distribution plot
sns.distplot(x=dfnew['symboling'],color='hotpink')
```

```python
sns.distplot(x=dfnew['symboling'],color='hotpink',kde=False)
#without kernel density estimator---->kde
```

```python
sns.jointplot('engine_size','fuel_type',data=dfnew)
plt.show()
#jointplot is a bivariate plot along with distribution plot
```

```python
sns.violinplot(x='compression_ratio',y='body_style',data=dfnew)
plt.xticks(color='m')
plt.yticks(color='m')
#similar to boxplot
#consists of kde
```

```python
sns.violinplot(x='compression_ratio',y='body_style',data=dfnew,hue='fuel_type',split=True)
plt.xticks(color='m')
plt.yticks(color='m')
plt.xlabel('compression',color='olive')
plt.ylabel('body_style',color='olive')
```

```python
#swarmplots
#combination of split and violin plots
```

```python
sns.swarmplot(y='highway_mpg',x='fuel_system',data=dfnew)
```

```python
dfnew
```

```python
sns.swarmplot(y='highway_mpg',x='fuel_system',hue='engine_location',data=dfnew)
plt.legend()
```

```python
sns.regplot('wheel_base','engine_size',data=dfnew)
#regression plot
```
